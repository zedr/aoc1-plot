```python
from matplotlib import pyplot as plt

from numlib import abs_diff, split_text_num
```

Processo il file:

```python
fd = open("input.txt")
```

Prima importo la funzione per separare i due numeri:

```python
y = []

for line in fd:
    a, b = split_text_num(line)
    y.append(abs_diff(a, b))
```

```python
plt.plot(y)
plt.show()
```
